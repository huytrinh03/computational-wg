```python
import pandas as pd
import re

CSV_FILE_PATH = "foundit-IT-cleaned.csv"

df = pd.read_csv(CSV_FILE_PATH)

prompts = []

for index, row in df.iterrows():
    prompt = "I am conducting a correspondance study to investigate the role of gender discrimination in restricting women's labor force participation in India. I need to create many sets of 2 CVs that are identical except for the name, where one CV has a male-sounding Indian name and the other CV has a female-sounding Indian name. I would like that the CVs in each set are targeted to a specific job posting. The CVs must have the following sections: Personal information (name, email, city), Education, and Work Experience. Please create a one-page CV in English for a job posting with the following characteristics:"
    for column_name, value in row.items():
        # Clean the column_name by replacing any non-alphanumeric value with a space
        column_name = re.sub("\W", " ", column_name)
        prompt += " {}: {}.".format(column_name, value)
    prompts.append(prompt)

df.insert(len(df.columns), "gen_ai_prompt", prompts)

df["gen_ai_prompt"][4]
```
